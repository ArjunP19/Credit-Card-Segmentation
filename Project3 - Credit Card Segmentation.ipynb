# Outlier Detection

```python
#Outlier detection

plt.figure(figsize=(8,100))
for n, i in enumerate(df_features):
    n=n+1
    plt.subplot(17,1,n)
    sns.boxplot(ccard_df[i])
```

```python
#Identify the outlier in dataset from IQR method
Q25 = ccard_df.quantile(0.25)
Q75 = ccard_df.quantile(0.75)
IQR = Q75 - Q25

lower_bound = Q25 - (1.5 * IQR)
upper_bound = Q75 + (1.5 * IQR)
```

```python
#extract the columns names of outlier exists for
name=list()
for n in ccard_df.columns:
    if n in lower_bound.index:
        name.append(n)
```

```python
ccard_df=ccard_df[name]
```

```python
#copy the outliers into a dataframe
outlier_df=ccard_df[~((ccard_df < lower_bound) | (ccard_df > upper_bound)).any(axis=1)]
```

```python
#Dispolay the outlier dataframe
outlier_df.head()
```

> 2.Purchased by Type (one-off, instalments)

```python
#People not done both type of purchases
ccard_df[(ccard_df['ONEOFF_PURCHASES']==0) & (ccard_df['INSTALLMENTS_PURCHASES']==0)].shape
```

4 types of purchase behaviour has been seen
         1. People not done both type of purchases (one-off, instalments)
         2. People done both type of purchases (one-off, instalments)
         3. People done only one-off payment type of purchases
         4. People done only installments payment type of purchases

```python
def purchase(ccard_df):
    if (ccard_df['ONEOFF_PURCHASES']==0) & (ccard_df['INSTALLMENTS_PURCHASES']==0):
        return 'p_none'
    if (ccard_df['ONEOFF_PURCHASES']>0) & (ccard_df['INSTALLMENTS_PURCHASES']>0):
        return 'p_both'
    if (ccard_df['ONEOFF_PURCHASES']>0) & (ccard_df['INSTALLMENTS_PURCHASES']==0):
        return 'p_oneoff_installment'
    if (ccard_df['ONEOFF_PURCHASES']==0) & (ccard_df['INSTALLMENTS_PURCHASES']>0):
        return 'p_installment'
    
```

```python
ccard_df['PURCHASE_TYPE']= ccard_df.apply(purchase, axis=1)
```

```python
#Counts Purchase type
ccard_df['PURCHASE_TYPE'].value_counts()
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca_series={}
N=range(3,15)
for n in N:
    pca = PCA(n_components=n)
    pca_variance = pca.fit(credit_scaled_df)
    pca_series[n]=sum(pca_variance.explained_variance_ratio_)
```

```python
pca_series
```

```python
pd.Series(pca_series).plot()
```

```python
pca_df=PCA(n_components=7).fit(credit_scaled_df)
pca_fit_df=pca_df.fit_transform(credit_scaled_df)
```

# K-mean clustering

```python
from sklearn.cluster import KMeans
```

```python
distoration=[]
K= range(1,10)
for n in K:
    Kmean_model=KMeans(n_clusters=n)
    Kmean_model.fit(pca_fit_df)
    distoration.append(Kmean_model.inertia_)
```

```python
#Plot KMean graph called Elbow method
plt.plot(K,distoration, 'bx-')
plt.xlabel('K')
plt.ylabel('Distoration')
plt.title("KMean graph using ELBOW method to know optimal K")
```

# Checking performance metrics for Kmeans

```python
from sklearn.metrics import calinski_harabaz_score,silhouette_score
```

```python
SS_5 = silhouette_score(pca_fit_df,km_5.labels_)
CHS_5 = calinski_harabaz_score(pca_fit_df,km_5.labels_)
SS_6 = silhouette_score(pca_fit_df,km_6.labels_)
CHS_6 = calinski_harabaz_score(pca_fit_df,km_6.labels_)
SS_7 = silhouette_score(pca_fit_df,km_7.labels_)
CHS_7 = calinski_harabaz_score(pca_fit_df,km_7.labels_)

print("Clusters            -  silhouette_score   -  calinski_harabaz_score")
print("Kmean - 5 clusters :- ", SS_5, "-" , CHS_5)
print("Kmean - 6 clusters :- ", SS_6, "-" , CHS_6)
print("Kmean - 7 clusters :- ", SS_7, "-" , CHS_7)
```

```python
score_ch={}
score_s={}
for n in range(3,10):
    km_score=KMeans(n_clusters=n)
    km_score.fit(pca_fit_df)
    score_s[n]=calinski_harabaz_score(pca_fit_df,km_score.labels_)
    score_ch[n]=silhouette_score(pca_fit_df,km_score.labels_)
```
